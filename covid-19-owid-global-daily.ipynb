## 可視化

```python
import matplotlib.pyplot as plt
#%matplotlib notebook
%matplotlib inline
```

```python
min_date = pd.to_datetime("2021-05-01")
```

```python
world_df["case_rolling7ave"] = world_df.new_cases.rolling(7).mean()
world_df[["new_cases_smoothed","case_rolling7ave"]]
```

```python
fig=plt.figure(figsize=(20,12))
plt.subplot(2,2,1) # (rows, columns, panel number)
plt.title("Daily new confirmed COVID-19 cases in the world")
plt.bar(world_df.date,world_df.new_cases,label="New cases")
plt.plot_date(world_df["date"],world_df["new_cases_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
#plt.xticks(rotation=70)
plt.ylabel("people")
plt.grid(axis="y")
plt.legend()

plt.subplot(2,2,2) # (rows, columns, panel number)
plt.title("Daily new confirmed COVID-19 cases in the world(Recent)")
chart_data = world_df[world_df.date >= min_date]
plt.bar(chart_data["date"],chart_data["new_cases"],label="New cases")
plt.plot_date(chart_data["date"],chart_data["new_cases_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
#plt.xticks(rotation=70)
plt.grid(axis="y")
plt.legend()

plt.subplot(2,2,3) # (rows, columns, panel number)
plt.title("Daily new deaths in the world")
plt.bar(world_df.date,world_df.new_deaths,label="New deaths")
plt.plot_date(world_df["date"],world_df["new_deaths_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
#plt.xticks(rotation=70)
plt.ylabel("people")
plt.grid(axis="y")
plt.legend()

plt.subplot(2,2,4) # (rows, columns, panel number)
plt.title("Daily new deaths in the world(Recent)")
chart_data = world_df[world_df.date >= min_date]
plt.bar(chart_data["date"],chart_data["new_deaths"],label="New deaths")
plt.plot_date(chart_data["date"],chart_data["new_deaths_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
#plt.xticks(rotation=70)
plt.grid(axis="y")
plt.legend()

fig.text(0,1,"Latest: {}".format(latest_date_str),ha="left",va="top")
fig.text(0,0
         ,"*Data Source : COVID-19 Dataset by Our World in Data",
        ha="left",va="bottom")

fig.tight_layout()
```

### 新規感染者数

```python
fig = plt.figure(figsize=(12,14))

plt.subplot(2,2,1) # (rows, columns, panel number)
plt.title("New confirmed COVID-19 cases worst 50")
plt.yticks(fontsize=9)
plt.xlabel("people")
chart_data = c_df_latest.sort_values("new_cases",ascending=False).head(50).sort_values("new_cases",ascending=True)
plt.barh(chart_data["location"],chart_data["new_cases"],label="new cases a day")
plt.grid(axis="x")
mean = int(c_df_latest.new_cases.mean())
plt.plot([mean,mean],[0,50], 'k--', lw=0.5)
plt.text(mean,50,"Mean:{}".format(mean),ha="center",fontsize=9)
plt.legend()

plt.subplot(2,2,2) # (rows, columns, panel number)
plt.title("New confirmed COVID-19 cases per million worst 50")
plt.yticks(fontsize=9)
plt.xlabel("people")
chart_data = c_df_latest.sort_values("new_cases_per_million",ascending=False).head(50).sort_values("new_cases_per_million",ascending=True)
plt.barh(chart_data["location"],chart_data["new_cases_per_million"],label="new cases a day per million")
plt.grid(axis="x")
mean = int(chart_data.new_cases_per_million.mean())
plt.plot([mean,mean],[0,50], 'k--', lw=0.5)
plt.text(mean,50,"Mean:{}".format(mean),ha="center",fontsize=9)
plt.legend()

fig.text(0,0.98,"Latest: {}".format(latest_date_str),fontsize=9,ha="left",va="top")
fig.text(0,0
         ,"*Data Source : COVID-19 Dataset by Our World in Data",
        fontsize=8,ha="left",va="bottom")

plt.subplot(2,2,3) # (rows, columns, panel number)
plt.title("New deaths worst 50")
plt.yticks(fontsize=9)
plt.xlabel("people")
chart_data = c_df_latest.sort_values("new_deaths",ascending=False).head(50).sort_values("new_deaths",ascending=True)
plt.barh(chart_data["location"],chart_data["new_deaths"],label="new deaths a day")
plt.grid(axis="x")
mean = int(c_df_latest.new_deaths.mean())
plt.plot([mean,mean],[0,50], 'k--', lw=0.5)
plt.text(mean,50,"Mean:{}".format(mean),ha="center",fontsize=9)
plt.legend()

plt.subplot(2,2,4) # (rows, columns, panel number)
plt.title("Total deaths per capita worst 50")
plt.yticks(fontsize=9)
plt.xlabel("%")
chart_data = c_df_latest.copy()
chart_data["total_deaths_per_capita"] = chart_data.total_deaths / chart_data.population * 100
chart_data = chart_data.sort_values("total_deaths_per_capita",ascending=False).head(50).sort_values("total_deaths_per_capita",ascending=True)
plt.barh(chart_data["location"],chart_data["total_deaths_per_capita"],label="total deaths per capita")
plt.grid(axis="x")
mean = round(chart_data.total_deaths_per_capita.mean(),2)
plt.plot([mean,mean],[0,50], 'k--', lw=0.5)
plt.text(mean,50,"Mean:{}".format(mean),ha="center",fontsize=9)
plt.legend()


fig.tight_layout()
```
