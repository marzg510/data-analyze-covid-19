```python
import pandas as pd
```

```python
pref = pd.read_csv("COVID-19-jp/nhk_news_covid19_prefectures_daily_data.csv")
print(len(pref))
pref.head()
```

```python
pref["日付"] = pd.to_datetime(pref["日付"] )
```

```python
all_jp = pref.groupby(["日付"], as_index=False).sum()
del all_jp["都道府県コード"]
del all_jp["各地の感染者数_累計"]
del all_jp["各地の死者数_累計"]
all_jp.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
#%matplotlib inline
```

```python
all_jp
```

```python
plt.figure(figsize=(8,5))
plt.plot_date(all_jp["日付"],all_jp["各地の感染者数_1日ごとの発表数"],label="positive", linestyle="solid",marker="")
plt.plot_date(all_jp["日付"],all_jp["各地の死者数_1日ごとの発表数"],label="death", linestyle="solid",marker="")
plt.legend()
```
