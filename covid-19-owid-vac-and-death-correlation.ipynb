```python
import pandas as pd
```

```python
df = pd.read_csv("covid-19-data/public/data/owid-covid-data.csv")
df.head()
```

```python
df[df.continent.isna()].location.unique()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
# データを国だけにする
cdf = df.dropna(subset=["continent"])
```

```python
cdf[cdf.location=="Japan"]
```

```python
# ワクチン接種数データがない行を消す
cdf = cdf.dropna(subset=["people_fully_vaccinated"])
```

```python
# データを2021-01-01以降だけにする
cdf = cdf[cdf.date >= "2021-01-01"]
deaths = cdf.groupby("location", as_index=False).agg({"new_deaths":"sum"}).rename(columns={"new_deaths":"total_deaths"})
deaths = deaths[deaths.total_deaths>0]
deaths = deaths.dropna(subset=["total_deaths"])
deaths
```

```python
deaths[deaths.location=="Japan"]
```

```python
countries = cdf.groupby("location", as_index=False).agg({"date":"max"})
countries
```

```python
countries[countries.location=="Japan"]
```

```python
tdf = pd.merge(countries, cdf, on=["location","date"], how="inner")[["continent","location","date","people_fully_vaccinated","population"]]
tdf
```

```python
tdf[tdf.location=="Japan"]
```

```python
tdf = pd.merge(tdf, deaths, on=["location"], how="inner")
tdf["total_deaths_per_capita"] = tdf.total_deaths / tdf.population * 100
tdf["fully_vaccin_per_capita"] = tdf.people_fully_vaccinated / tdf.population * 100
tdf
```

```python
tdf[tdf.location=="United States"]
```

## 可視化

```python
import matplotlib.pyplot as plt
#%matplotlib notebook
%matplotlib inline
```

```python
continents = {
    'Africa':('b'),
    'South America':('g'),
    'Oceania':('r'),
    'Asia':('c'),
    'Europe':('y'),
    'North America':('m'),
}
```

```python
fig=plt.figure(figsize=(12,8))
#plt.subplot(2,2,1) # (rows, columns, panel number)
plt.title("COVID-19 Scatter of Total Vaccination and Total Deaths(2021.01.01 - {})".format(tdf.date.max().strftime("%Y.%m.%d")))
for c in continents:
    c_color = continents[c][0]
    plt.scatter(tdf[tdf.continent==c].fully_vaccin_per_capita, tdf[tdf.continent==c].total_deaths_per_capita,color=c_color,label=c)
for i,r in tdf.iterrows():
    plt.text(r.fully_vaccin_per_capita, r.total_deaths_per_capita,r.location,ha="center",va="bottom")
#plt.xticks(rotation=70)
plt.text(0,tdf.total_deaths_per_capita.max(),"*Excludes countries with no deaths",fontsize=10,ha="left",va="bottom")
plt.xlabel("fully vaccinations per capita(%)")
plt.ylabel("total deaths per capita(%)")
#plt.grid(axis="y")
plt.legend()
fig.text(0,0
         ,"*Data Source : COVID-19 Dataset by Our World in Data",
        fontsize=10,ha="left",va="bottom")

fig.tight_layout()
```
