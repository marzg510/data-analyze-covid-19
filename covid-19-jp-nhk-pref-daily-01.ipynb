```python
import pandas as pd
```

```python
pref = pd.read_csv("COVID-19-jp/nhk_news_covid19_prefectures_daily_data.csv")
print(len(pref))
pref.head()
```

```python
pref["日付"] = pd.to_datetime(pref["日付"] )
```

```python
saitama = pref.loc[pref["都道府県名"]=="埼玉県"]
saitama.tail(3)
```

```python
tokyo = pref.loc[pref["都道府県名"]=="東京都"]
tokyo.tail(3)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
#%matplotlib inline
```

```python
plt.figure(figsize=(8,5))
#plt.yscale("log")
#plt.plot(saitama["日付"].dt.strftime("%Y%m%d"),saitama["各地の感染者数_1日ごとの発表数"],label=saitama["都道府県名"].unique())
#plt.plot(saitama["日付"].dt.strftime("%Y%m%d"),saitama["各地の感染者数_1日ごとの発表数"],label="saitama")
plt.plot_date(saitama["日付"],saitama["各地の感染者数_1日ごとの発表数"],label="saitama", linestyle="solid",marker="")
#plt.plot(tokyo["日付"].dt.strftime("%Y%m%d"),tokyo["各地の感染者数_1日ごとの発表数"],label="tokyo")
plt.plot_date(tokyo["日付"],tokyo["各地の感染者数_1日ごとの発表数"],label="tokyo", linestyle="solid",marker="")
plt.legend()
```
