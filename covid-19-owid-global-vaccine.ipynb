```python
import pandas as pd
```

```python
df = pd.read_csv("covid-19-data/public/data/vaccinations/vaccinations.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["date"] = pd.to_datetime(df["date"])
df.info()
```

```python
import datetime
latest_date_str = (df.date.max() - datetime.timedelta(days=1)).strftime("%Y-%m-%d")
latest_date_str
```

## 全世界のデータ

```python
world_df = df.copy().loc[df.location=="World"]
world_df.tail()
```

## 国別

```python
country_df = df.loc[df.iso_code.str.len()==3]
country_df.head()
```

```python
latest_df = country_df.loc[country_df.date==latest_date_str]
latest_df.head()
```

### ワクチン接種完了率トップ20

```python
latest_df.sort_values(by="people_fully_vaccinated_per_hundred",ascending=False).head(20)[["location","iso_code","date","people_fully_vaccinated_per_hundred"]]
```

### ワクチン接種完了率ワースト20

```python
latest_df.sort_values(by="people_fully_vaccinated_per_hundred",ascending=True).head(20)[["location","iso_code","date","people_fully_vaccinated_per_hundred"]]
```
