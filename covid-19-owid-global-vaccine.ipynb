## 全世界のデータ

```python
world_df = df.copy().loc[df.location=="World"]
world_df.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
#%matplotlib inline
```

```python
fig=plt.figure(figsize=(12,8))
ax1 = plt.subplot(1,1,1) # (rows, columns, panel number)
plt.title("Daily COVID-19 vacctinated in the world")
ax1.set_zorder(2)
ax1.patch.set_alpha(0)
plt.ylim(0,100)
plt.ylabel("vaccination %")
ax1.grid()
plt.xticks(rotation=70)
ax2 = plt.twinx()
ax2.set_zorder(1)
plt.ylabel("million people")
ax1.plot_date(world_df["date"],world_df["people_vaccinated_per_hundred"],label="Vaccinated at least once per population", fmt="-")
ax1.plot_date(world_df["date"],world_df["people_fully_vaccinated_per_hundred"],label="Fully vaccinated per population", fmt="-")
ax2.bar(world_df["date"],world_df["people_vaccinated"]/1000/1000,label="People vaccinated at least once", color="pink")
ax2.bar(world_df["date"],world_df["people_fully_vaccinated"]/1000/1000,label="People fully vaccinated", color="lightgray",zorder=1)
#plt.grid(axis="y")
h1,l1 = ax1.get_legend_handles_labels()
h2,l2 = ax2.get_legend_handles_labels()
plt.legend(h1+h2,l1+l2)
"""
plt.subplot(2,2,2) # (rows, columns, panel number)
plt.title("Daily new confirmed COVID-19 cases in the world(after 2021 Apr 1st)")
chart_data = world_df[world_df.date >= "2021-04-01"]
plt.bar(chart_data["date"],chart_data["new_cases"],label="New deaths")
plt.plot_date(chart_data["date"],chart_data["new_cases_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
plt.xticks(rotation=70)
plt.grid(axis="y")
plt.legend()

plt.subplot(2,2,3) # (rows, columns, panel number)
plt.title("Daily new deaths in the world")
plt.bar(world_df.date,world_df.new_deaths,label="New cases")
plt.plot_date(world_df["date"],world_df["new_deaths_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
plt.xticks(rotation=70)
plt.ylabel("people")
plt.grid(axis="y")
plt.legend()

plt.subplot(2,2,4) # (rows, columns, panel number)
plt.title("Daily new deaths in the world(after 2021 Apr 1st)")
chart_data = world_df[world_df.date >= "2021-04-01"]
plt.bar(chart_data["date"],chart_data["new_deaths"],label="New deaths")
plt.plot_date(chart_data["date"],chart_data["new_deaths_smoothed"],label="rolling 7-day average", fmt="-", color="orange")
plt.xticks(rotation=70)
plt.grid(axis="y")
plt.legend()
"""

fig.text(0,1,"Latest: {}".format(latest_date_str),fontsize=8,ha="left",va="top")
fig.text(0,0
         ,"*Data reference from:COVID-19 Dataset by Our World in Data",
        fontsize=7,ha="left",va="bottom")

fig.tight_layout()
```
